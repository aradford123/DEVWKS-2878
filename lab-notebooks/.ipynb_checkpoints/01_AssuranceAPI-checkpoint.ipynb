### Sync response
In order to get an immediate response, a header "__runsync" is required.

This request give the network-health during a set of time.  The startTime and endTime need to be specified even if they are empty.

```python
response = get_url("dna/intent/api/v1/network-health?startTime=&endTime=",
                   headers={'__runsync': 'true'})
```

```python
print json.dumps(response,indent=2)
```

### site-hierarchy
The next API gets a site hierarcy.  Again, timestamp is required, but can be empty.


```python
response = get_url("dna/intent/api/v1/site-hierarchy?timestamp=",
                   headers={'__runsync': 'true'})
```

```python
print json.dumps(response,indent=2)
```

```python
print "\n".join(["Site:{}, NwkHealth:{}".format(res['siteName'], 
                      res['networkHealthAccess'])
                for res in response['response']])
```

```python
timestamp=int(time.time() * 1000)
mac="00:1E:13:A5:B9:40"
url='dna/intent/api/v1/client-detail?timestamp={}&macAddress={}'.format(timestamp,mac)
response = get_url(url, headers={'__runsync': 'true'})
```

```python
print json.dumps(response,indent=2)
```

```python
timestamp = response['response']['connectionInfo']['timestamp']
print time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(timestamp/1000))
```
